```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
reservoir=pd.read_csv("chennai_reservoir_levels.csv",parse_dates=['Date'])
rainfall=pd.read_csv("chennai_reservoir_rainfall.csv",parse_dates=['Date'])
rainfall.set_index('Date',inplace=True)
reservoir.set_index('Date',inplace=True)
print(reservoir)
print(rainfall)
```

```python
print("Highest water level in Poondi reservoir: ",reservoir['POONDI'].max())
print("Highest water level in Cholavaram reservoir: ",reservoir['CHOLAVARAM'].max())
print("Highest water level in Redhills reservoir: ",reservoir['REDHILLS'].max())
print("Highest water level in Chembarambakkam reservoir: ",reservoir['CHEMBARAMBAKKAM'].max())
```

```python
print("Lowest water level in Poondi reservoir: ",reservoir['POONDI'].min())
print("Lowest water level in Cholavaram reservoir: ",reservoir['CHOLAVARAM'].min())
print("Lowest water level in Redhills reservoir: ",reservoir['REDHILLS'].min())
print("Lowest water level in Chembarambakkam reservoir: ",reservoir['CHEMBARAMBAKKAM'].min())
```

```python
print("Highest rainfall in Poondi reservoir: ",rainfall['POONDI'].max())
print("Highest rainfall in Cholavaram reservoir: ",rainfall['CHOLAVARAM'].max())
print("Highest rainfall in Redhills reservoir: ",rainfall['REDHILLS'].max())
print("Highest rainfall in Chembarambakkam reservoir: ",rainfall['CHEMBARAMBAKKAM'].max())
```

```python
print("Lowest rainfall in Poondi reservoir: ",rainfall['POONDI'].min())
print("Lowest rainfall in Cholavaram reservoir: ",rainfall['CHOLAVARAM'].min())
print("Lowest rainfall in Redhills reservoir: ",rainfall['REDHILLS'].min())
print("Lowest rainfall in Chembarambakkam reservoir: ",rainfall['CHEMBARAMBAKKAM'].min())
```
